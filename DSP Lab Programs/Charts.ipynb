```python
#Bar chart
import matplotlib.pyplot as plt
Country=["India", "Korea", "Japan", "UK", "Germany","France"]
GDP_Per_Capita=[48000, 50000, 55000, 35000, 60000, 25000]
plt.bar(Country, GDP_Per_Capita)
plt.title('Country vs GDP Per Capita')
plt.xlabel('Country')
plt.ylabel('GDP Per capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
Country=["India", "Korea", "Japan", "UK", "Germany","France"]
GDP_Per_Capita=[48000, 50000, 55000, 35000, 60000, 25000]
plt.barh(Country, GDP_Per_Capita)
plt.title('Country vs GDP Per Capita')
plt.xlabel('Country')
plt.ylabel('GDP Per capita')
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
numbers=[10, 20 ,30, 40, 50, 60]
plt.hist(numbers,bins=5, color="pink")
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```

```python
#pie
import numpy as np
y=np.array([45,55,95])
co=["r", "b", "hotpink"]
plt.pie(y,colors=co, labels=["Female", "Male", "Others"])
plt.show()
```
